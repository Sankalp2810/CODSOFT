# Credit Card Fraud Detection using Logistic Regression

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```
card_data = pd.read_csv("creditcard.csv")
card_data.head()
```

```
card_data.info()
card_data.isnull().sum()
card_data["Class"].value_counts()
```

```
safe=card_data[card_data.Class==0]
fraud=card_data[card_data.Class==1]
safe_sample=safe.sample(n=492, random_state=2)
newdata=pd.concat([safe_sample,fraud],axis=0)
newdata["Class"].value_counts()
```

```
X=newdata.drop(columns="Class",axis=1)
Y=newdata["Class"]

X_train,X_test,Y_train,Y_test=train_test_split(
    X,Y,test_size=0.2,stratify=Y,random_state=2)

model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)

print("Training Accuracy:",accuracy_score(Y_train,model.predict(X_train)))
print("Testing Accuracy:",accuracy_score(Y_test,model.predict(X_test)))
```
